Solution

```python
##### Pb.2(a): solutions
import numpy as np
from numpy import linalg as LA
# ensure 'ncon.py' in working dir
from ncon import ncon

##### Define matrix A
d1 = 10
d2 = 8
A = np.zeros((d1,d1,d2,d2))
for ni in range(d1):
    for nj in range(d1):
        for nk in range(d2):
            for nl in range(d2):
                A[ni,nj,nk,nl] = np.sqrt((ni+1)+2*(nj+1)+3*(nk+1)+4*(nl+1))
            
##### Compute norm and normalize A
n_A = LA.norm(A)
Ap = A / n_A

##### take SVD of A
Um,Sm,Vmh = LA.svd(Ap.reshape(d1**2,d2**2))
m_A = np.sqrt(sum(Sm**2)) #should equal 1

##### effective rank
deltaval = 1e-4
r_delta = sum(Sm > deltaval)

##### truncation error
err_0 = np.sqrt(sum((Sm[r_delta:])**2))

##### optimal restricted rank approximation
B = Um[:,:r_delta] @ np.diag(Sm[:r_delta]) @ Vmh[:r_delta,:]
B = B.reshape(d1,d1,d2,d2)
err_1 = LA.norm(Ap - B)

# compare 
[err_0,err_1]
```
